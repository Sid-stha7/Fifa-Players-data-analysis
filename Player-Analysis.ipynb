# Players With ratings higher than 90

```python
overall_rating=df[['Name','Overall']]
overall_rating.head(15)
```

```python
highest_rating=overall_rating.sort_values(by=['Overall'],ascending=False)
highest_rating
```

By this analysis it is confirmed that there are 10 players with rating higher or equal to 90

```python
over_90=highest_rating.loc[highest_rating['Overall'] >=90]
over_90
```

```python
x=list(over_90['Name'].value_counts()[0:10].keys())
y=over_90['Overall'][0:10]
y
```

```python
plt.figure(figsize=(15,5))
plt.bar(x, y, color =['darkblue'])
plt.xlabel("Player Names")
plt.ylabel("Over All Ratings")
```
